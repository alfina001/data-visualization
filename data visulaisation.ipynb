```python
data['Gender'].unique()
```

```python
data['Gender']=data['Gender'].replace(to_replace=['Male','Female'],value=[0,1])
```

```python
data['Gender']=data['Gender'].fillna(0)
```

```python
female=data['Gender'].value_counts()[1]
male=data['Gender'].value_counts()[0]
x=[female,male]
```

```python
plt.pie(x,labels=['female','male'],explode=[0,0.1],autopct='%1.1f%%')
```

```python
plt.scatter(df['height'],df['age'],marker='o')
plt.scatter(df['weight'],df['age'],marker='o')

plt.plot(df['height'],df['age'],linestyle='--')
plt.plot(df['weight'],df['age'],linestyle='--')

plt.title('line chart')
plt.xlabel('height')
plt.ylabel('age')
```

```python
plt.hist(df['height'])
```

```python
data['Property_Area'].unique()
```

```python
data['Property_Area']=data['Property_Area'].replace(to_replace=['Urban', 'Rural', 'Semiurban'],value=[0,1,2])
```

```python
Rural=data['Property_Area'].value_counts()[1]
Urban=data['Property_Area'].value_counts()[0]
Semiurban=data['Property_Area'].value_counts()[2]
x=[Rural,Urban,Semiurban]
```

```python
plt.pie(x,labels=['Rural','Urban','Semiurban'],explode=[0,0.1,0.1],autopct='%1.1f%%')
```
